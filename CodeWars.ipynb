```python
def solution(number):
    lst = []
    if number < 0:
        return 0
    for n in range(1, number):
        if n % 3 == 0 or n % 5 == 0:
            lst.append(n)
    return sum(lst)


solution(10)
```

```python
def is_square(n):
    root = str(n ** (1/2))
    if n == 0:
        return True
    elif (root)[-1] == '0' and (root)[-2] == '.':
        return True
    else:
        return False

is_square(5)
```

```python
def accum(s):
    letters = list(s)
    l = []
    for i, c in enumerate(letters):
#         print(i, c)
        l.append(((i+1)*c).title())
    return '-'.join(l)
    
# accum("abcd") # -> "A-Bb-Ccc-Dddd"
# accum("RqaEzty") # -> "R-Qq-Aaa-Eeee-Zzzzz-Tttttt-Yyyyyyy"
accum("cwAt") # -> "C-Ww-Aaa-Tttt"    
```

```python
def get_middle(s):
    if len(s) % 2 == 0:
        return ''.join([s[int(len(s)/2)-1], s[int(len(s)/2)]])
    else:
        return s[int(len(s)/2)]
    
get_middle('odemaodemas')
```

```python
def descending_order(num):
    
    string = str(num)
    l = []
    for n in string:
        l.append(n)
        
    l.sort(reverse=True)
    return int(''.join(l))

descending_order(42145)
```

```python
def abbrev_name(name):
    initials = ""
    names = name.split()
    for word in names:
        initials += word[0].upper() + "."
    return (initials.strip("."))
    
    
abbrev_name('sam harris')
```

```python
def find_needle(haystack):
    
    for junk in range(len(haystack)):
        if haystack[junk].lower() == 'needle':
            return f"found the needle at position {junk}"
        
find_needle(['1', 'needle'])
```

```python
def divisors(n):
    divs = []
    for num in range(1, n + 1):
        if n % num == 0:
            print(num)
            divs.append(num)
    return len(divs)

divisors(30)

# 4 --> 3 (1, 2, 4)
# 5 --> 2 (1, 5)
# 12 --> 6 (1, 2, 3, 4, 6, 12)
# 30 --> 8 (1, 2, 3, 5, 6, 10, 15, 30)
```
